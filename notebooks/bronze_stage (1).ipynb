```python
from pyspark.sql.functions import * 
from pyspark.sql.types import *
```

```python
patients_path="abfss://hospital_data_pipeline@onelake.dfs.fabric.microsoft.com/hospital_lakehouse.Lakehouse/Files/raw/patients/patients.csv"
patient_schema=StructType([
    StructField("patient_id",IntegerType()),
    StructField("age",IntegerType()),
    StructField("gender",StringType()),
    StructField("comorbidity_index",IntegerType()),
    StructField("admission_type", StringType()),
    StructField("baseline_risk_score", DoubleType()),
    StructField("los_hours", IntegerType()),
    StructField("deterioration_event", IntegerType()),
    StructField("deterioration_within_12h_from_admission", IntegerType()),
    StructField("deterioration_hour", IntegerType())


])
df_patients=spark.read.schema(patient_schema).option("header",True).csv(patients_path).withColumn("ingestion_timestamp", current_timestamp())
df_patients.write.format("delta").mode("overwrite").saveAsTable("bronze_patients")
```

```python
labs_path="abfss://hospital_data_pipeline@onelake.dfs.fabric.microsoft.com/hospital_lakehouse.Lakehouse/Files/raw/labs/labs_timeseries.csv"
labs_schema=StructType([StructField("patient_id", IntegerType()),
   StructField("hour_from_admission", IntegerType()),
    StructField("wbc_count", DoubleType()),
    StructField("lactate", DoubleType()),
    StructField("creatinine", DoubleType()),
    StructField("crp_level", DoubleType()),
    StructField("hemoglobin", DoubleType()),
    StructField("sepsis_risk_score", DoubleType())

])
df_labs=spark.read.schema(labs_schema).option("header",True).csv(labs_path).withColumn("ingestion_timestamp", current_timestamp())
df_labs.write.format("delta").mode("overwrite").saveAsTable("bronze_labs")
```

```python
vital_path="abfss://hospital_data_pipeline@onelake.dfs.fabric.microsoft.com/hospital_lakehouse.Lakehouse/Files/raw/vitals/vitals_timeseries.csv"
vital_schema=StructType([


     StructField("patient_id", IntegerType()),
    StructField("hour_from_admission", IntegerType()),
    StructField("heart_rate", DoubleType()),
    StructField("respiratory_rate", DoubleType()),
    StructField("spo2_pct", DoubleType()),
    StructField("temperature_c", DoubleType()),
    StructField("systolic_bp", DoubleType()),
    StructField("diastolic_bp", DoubleType()),
    StructField("oxygen_device", StringType()),
    StructField("oxygen_flow", DoubleType()),
    StructField("mobility_score", IntegerType()),
    StructField("nurse_alert", IntegerType())
])

df_vitals=spark.read.schema(vital_schema).option("header",True).csv(vital_path).withColumn("ingestion_timestamp", current_timestamp())
df_vitals.write.format("delta").mode("overwrite").saveAsTable("bronze_vitals")
```
